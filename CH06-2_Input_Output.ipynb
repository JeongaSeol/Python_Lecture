```python
f = open("two_times_table.txt",'w')
for num in range(1,6):
    format_string = "2 * {0} = {1}\n".format(num, 2*num)
    f.write(format_string)
f.close()
!cat two_times_table.txt
```

```python
f = open("n_times_table.txt",'w')
num = int(input("원하시는 단을 입력해주세요 :"))
for i in range(1,10) :
    string = "{0} * {1} = {2}\n".format(num, i, num*i)
    f.write(string)
f.close()
!cat n_times_table.txt
```

```python
f = open("times_table.txt", 'w')
for i in range(2, 10, 3) :
    for j in range(1, 10) :
        string2 = "{0} * {1} = {2}\t".format(i, j, i*j)
        string3 = "{0} * {1} = {2}\t".format(i+1, j, (i+1)*j)
        string4 = "{0} * {1} = {2}\n".format(i+2, j, (i+2)*j)
        f.write(string2)
        f.write(string3)
        f.write(string4)
    f.write("\n")
f.close()
!cat times_table.txt
```
