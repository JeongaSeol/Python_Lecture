```python
import datetime

set_day = datetime.date(2021,8,4)
print(set_day)
```

```python
print('{0}/{1}/{2}'.format(set_day.year, set_day.month, set_day.day))
```

```python
import datetime

day1 = datetime.date(2021,4,1)
day2 = datetime.date(2021,7,10)
diff_day = day2 - day1
print('day2 - day1 = ', diff_day)

print(type(day1))
print(type(diff_day))
```

```python
print("** 지정된 두 날짜의 차이는 {0}일입니다.".format(diff_day.days))
```

```python
today = datetime.date.today()
sp_day = datetime.date(2020,12,31)
print(today)
print('오늘은 {0}로 부터 {1}일째 되는 날입니다.'.format(sp_day,  (today-sp_day).days))
```
